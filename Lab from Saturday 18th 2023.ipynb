**8. Understand the meaning of all columns**

**customer:** identifier for each customer
**st:** state where the customer lives      
**gender:** gender (M = male, F = female)      
**education:** educational level of the customer    
**customer_lifetime_value:** total monetary value of the customer for the whole cycle    
**monthly_premium_auto:** amount the customer pays for insurance premium on a monthly basis   
**number_of_open_complaints:** the number of complaints that are currently open    
**policy_type:** type of insurance policy     
**vehicle_class:** class of the customer's vehicle    
**total_claim_amount:** the total amount claimed by the customer
